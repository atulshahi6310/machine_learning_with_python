```python
## elebow methods 
```

```python
from sklearn.cluster import KMeans

wcss = []

for k in range(1, 11):
    kmean = KMeans(
        n_clusters=k,
        init='k-means++',
        random_state=42
    )
    kmean.fit(x_train)
    wcss.append(kmean.inertia_)   #  FIXED
```

```python
wcss
```

```python
#plot the elbow curve 
plt.plot(range(1,11),wcss)
plt.xticks(range(1, 11))
```
