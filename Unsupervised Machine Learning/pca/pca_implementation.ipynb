```python
import matplotlib.pyplot as plt
import pandas  as pd
import numpy as np
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
df3 = load_breast_cancer()
```

```python
df3.keys()
```

```python
print(df3.DESCR)
```

```python
df = pd.DataFrame(df3["data"],columns=df3["feature_names"])
```

```python
df.head()
```

```python
##  pca is more abot dimensolality reduction
```

```python
## stanndardization 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_df = scaler.transform(df)
```

```python
scaled_df
```

```python
## applying pca allgorithm 
from sklearn.decomposition import PCA
```

```python
pca =PCA(n_components=2)
```

```python
df1 = pca.fit_transform(scaled_df) ## this is dimensolality reduction
```

```python
df1
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
plt.scatter(df1[:,0],df1[:,1],c=df3['target'] )
```
