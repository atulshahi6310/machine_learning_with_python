```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Books_Data_Clean.csv')
```

```python
df.head()
```

## cleaning the dataset

```python
df.describe()
```

```python
df =df[df['Publishing Year'] >1900]
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# we remove those book whose book name is null so we use pandas to dro the given book name feature null values 
df = df.copy()
df.dropna( subset="Book Name" , inplace= True)
```

```python
df.isnull().sum()
```

```python
# now cheak for is any row is duplicate exist or not using pandas 
df.duplicated().sum()
```

```python
sns.histplot(data=df, x="Publishing Year", kde=True)
plt.show()
```

```python
df.head(3)
```

```python
df['genre'].value_counts().plot(kind = 'bar')
plt.xlabel('genre')
plt.ylabel('number of books')
plt.title('number of book each genre')
plt.tight_layout()
```

```python
df.info()
```

```python
df.groupby('Author')['Book_average_rating'].mean().sort_values(ascending= False)
```

```python
sns.scatterplot(x = 'sale price' ,y ='units sold', data =df)
plt.title(" Sale Price vs unit sold")
plt.show()
```

```python
language_count = df["language_code"].value_counts()
```

```python
plt.figure(figsize=(5,5))
plt.pie(
    language_count,
    labels=language_count.index,
    startangle=90,
    autopct='%1.1f%%'
)
plt.title("Language Distribution")
plt.tight_layout()
plt.show()
```
