```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# DataFrame
df = pd.DataFrame({
    "color":['red','black','pink','red','blue','green']
})

# Create OneHotEncoder instance
encoder = OneHotEncoder()

# Fit and transform
encoded_array = encoder.fit_transform(df[["color"]]).toarray()

# Create dataframe with encoded columns
encoder_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(['color']))

# Concatenate original + encoded
final_df = pd.concat([df, encoder_df], axis=1)

final_df
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# DataFrame
df = pd.DataFrame({
    'cars' : ['pagani','astenmartin','lembo','bmw','tesla','byd','bmw']
})

encoder = OneHotEncoder()
encoder_array = encoder.fit_transform(df[['cars']]).toarray()
encoder_df = pd.DataFrame(encoder_array, columns=encoder.get_feature_names_out(['cars']))
finnal_df = pd.concat( [df, encoder_df] , axis= 1)
finnal_df
```
