```python
from sklearn.preprocessing import LabelEncoder

# sample data
colors = ['Red', 'Blue', 'Green', 'Blue', 'Red']

# create object
le = LabelEncoder()

# fit and transform
encoded_colors = le.fit_transform(colors)

print(encoded_colors)
print(le.classes_)
```

# ordinal encodeinng

```python
from sklearn.preprocessing import OrdinalEncoder
import numpy as np

data = np.array([
    ['Low'],
    ['Medium'],
    ['High'],
    ['Medium']
])

# define order of categories
categories = [['Low', 'Medium', 'High']]

oe = OrdinalEncoder(categories=categories)

encoded = oe.fit_transform(data)
print(encoded)
```

## traget guided  encodeing 

```python
import pandas as pd

df = pd.DataFrame({
    'City': ['Mumbai', 'Delhi', 'Bangalore', 'Hyderabad', 'Chennai',
             'Pune', 'Kolkata', 'Ahmedabad', 'Jaipur', 'Surat'],
    'Price': [75000, 68000, 72000, 60000, 65000,
              55000, 50000, 48000, 52000, 47000]
})

means = df.groupby('City')['Price'].mean().to_dict()
df['city_encoded'] =df['City'].map(means)
```
